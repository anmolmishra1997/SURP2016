```python
import csv
f1 = open ("C:\Users\Anmol Mishra\CC3DWorkspace\Phase 2\Problem 2\Circle R2\Demo_cc3d_05_21_2016_17_06_45\data.csv" , "rb")
data = csv.reader(f1)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
newdata = []
for line in data:
    newdata.append(line)
```

```python
tend = 596
```

```python
finalVolume = 0
for i in range(len(newdata)):
    if (int(newdata[i][3])>finalVolume ):
        finalVolume = int(newdata[i][3])
    if (int(newdata[i][0]) > tend + 100):
        break
```

```python
finalVolume
```

```python
freespace = []
for i in range(len(newdata)):
    if (int(newdata[i][0]) != int(newdata[i-1][0])):
        freespace.append(40000 - int(newdata[i][3]))
```

```python
for i in range(tend + 100):
    freespace[i] = freespace[i] + finalVolume - 40000
```

```python
plt.plot(freespace , linewidth = 2 , label = "R2")
plt.plot( [tend], freespace[tend] , 'ro' , label = "T_end")
plt.plot( [tend + 99], freespace[tend + 99] , 'go', label = "T_end + 100")
plt.title("Free space vs time")
plt.xlabel("MCS")
plt.ylabel("Free volume")
plt.legend()
plt.savefig("Free space Circle R2")
```

```python
noofcells = []
count =0
for i in range(len(newdata)):
    if (int(newdata[i][0]) == int(newdata[i-1][0])):
        count = count+1
    else :
        noofcells.append(count+1)
        count=0
```

```python
plt.plot( noofcells, linewidth = 2 , label = "Circle R2")
plt.plot( [tend], noofcells[tend] , 'ro' , label = "T_end")
plt.plot( [tend + 100], noofcells[tend + 100] , 'go', label = "T_end + 100")
plt.title("Number of cells vs time")
plt.xlabel("MCS")
plt.ylabel("Number of cells")
plt.legend( loc = 2)
plt.savefig("Number of Cells Circle R2")
```

```python
volume = []
varvolume = []
for i in range(len(newdata)):
    if (int(newdata[i][0]) == int(newdata[i-1][0]) and i != 0):
        volume.append(int(newdata[i-1][2]))
    else :
        volume.append(int(newdata[i-1][2]))
        npvolume = np.array(volume)
        varvolume.append(npvolume.std())
        volume = []
npvolume = np.array(volume)
varvolume.append(npvolume.std())
del varvolume[0]
```

```python
plt.plot(varvolume, label = "R2")
plt.plot( [tend], varvolume[tend] , 'ro' , label = "T_end")
plt.plot( [tend + 100], varvolume[tend + 100] , 'go', label = "T_end + 100")
plt.title("Volume heterogeneity vs time")
plt.xlabel("MCS")
plt.ylabel("Volume heterogeneity")
plt.legend(loc = 0)
plt.savefig("Volume Heterogeneity Circle R2")
```

```python
area = []
vararea = []
for i in range(len(newdata)):
    if (int(newdata[i][0]) == int(newdata[i-1][0]) and i != 0):
        area.append(float(newdata[i-1][4]))
    else :
        area.append(float(newdata[i-1][4]))
        nparea = np.array(area)
        vararea.append(nparea.std())
        area = []
nparea = np.array(area)
vararea.append(nparea.std())
del vararea[0]
```

```python
plt.plot(vararea, label = "R2")
plt.plot( [tend], vararea[tend] , 'ro' , label = "T_end")
plt.plot( [tend + 100], vararea[tend + 100] , 'go', label = "T_end + 100")
plt.title("Surface heterogeneity vs time")
plt.xlabel("MCS")
plt.ylabel("Surface heterogeneity")
plt.legend(loc = 4)
plt.savefig("Surface Heterogeneity Circle R2")
```
