```python
import csv
f1 = open ("C:\Users\Anmol Mishra\CC3DWorkspace\Phase 4\Circle R2\Demo.cc3d-2016-06-01-22-01-23-874000\data.csv" , "rb")
data = csv.reader(f1)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
newdata = []
for line in data:
    newdata.append(line)
finalVolume = int(newdata[-1][3])
```

```python
for i in range(len(newdata)):
    if (int(newdata[i][3])>finalVolume):
        finalVolume = int(newdata[i][3])
```

```python
tend = 0
for i in newdata:
    if (int(i[3]) == finalVolume) :
        tend = int(i[0])
        break
```

```python
tend
```

```python
freespace = []
for i in range(len(newdata)):
    if (int(newdata[i][0]) != int(newdata[i-1][0])):
        freespace.append(finalVolume - int(newdata[i][3]))
```

```python
plt.plot(freespace , linewidth = 2 , label = "Circle 2R")
plt.plot( [tend], freespace[tend] , 'ro' , label = "T_end")
plt.title("Free space vs time")
plt.xlabel("MCS")
plt.ylabel("Free volume")
plt.legend()
plt.savefig("Free space Circle 2R")
```

```python
noofcells = []
count =0
for i in range(len(newdata)):
    if (int(newdata[i][0]) == int(newdata[i-1][0])):
        count = count+1
    else :
        noofcells.append(count+1)
        count=0
```

```python
plt.plot( noofcells, linewidth = 2 , label = "Circle 2R")
plt.plot( [tend], noofcells[tend] , 'ro' , label = "T_end")
plt.title("Number of cells vs time")
plt.xlabel("MCS")
plt.ylabel("Number of cells")
plt.legend( loc = 2)
plt.savefig("Number of Cells Circle 2R")
```

```python
volume = []
varvolume = []
for i in range(len(newdata)):
    if (int(newdata[i][0]) == int(newdata[i-1][0]) and i != 0):
        volume.append(int(newdata[i-1][2]))
    else :
        volume.append(int(newdata[i-1][2]))
        npvolume = np.array(volume)
        varvolume.append(npvolume.std())
        volume = []
npvolume = np.array(volume)
varvolume.append(npvolume.std())
del varvolume[0]
```

```python
plt.plot(varvolume, label = "Circle 2R")
plt.plot( [tend], varvolume[tend] , 'ro' , label = "T_end")
plt.title("Volume heterogeneity vs time")
plt.xlabel("MCS")
plt.ylabel("Volume heterogeneity")
plt.legend(loc = 0)
plt.savefig("Volume Heterogeneity Circle 2R")
```

```python
area = []
vararea = []
for i in range(len(newdata)):
    if (int(newdata[i][0]) == int(newdata[i-1][0]) and i != 0):
        area.append(float(newdata[i-1][4]))
    else :
        area.append(float(newdata[i-1][4]))
        nparea = np.array(area)
        vararea.append(nparea.std())
        area = []
nparea = np.array(area)
vararea.append(nparea.std())
del vararea[0]
```

```python
plt.plot(vararea, label = "Circle 2R")
plt.plot( [tend], vararea[tend] , 'ro' , label = "T_end")
plt.title("Surface heterogeneity vs time")
plt.xlabel("MCS")
plt.ylabel("Surface heterogeneity")
plt.legend(loc = 4)
plt.savefig("Surface Heterogeneity Circle 2R")
```
