```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plottingdata = []
for i in datafresh:
    plottingdata.append(int(i[2]))
```

```python
def counter(x):
    i=0
    for j in plottingdata:
        if j==x:
            i=i+1
    return i
```

```python
y=[]
x = range(0,50)
for i in x:
    y.append(counter(i))
```
