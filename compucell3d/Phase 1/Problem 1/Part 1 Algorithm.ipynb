```python
import csv
f1 = open ("C:\Users\Anmol Mishra\CC3DWorkspace\Demo.cc3d-2016-05-17-09-43-51-863000\data.csv" , "rb")
data = csv.reader(f1)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
newdata = []
for line in data:
        newdata.append(line)
finalVolume = int(newdata[-1][3])
```

```python
tend = 0
for i in newdata:
    if (int(i[3]) == finalVolume) :
        tend = int(i[0])
        break
```

```python
tend
```

```python
freespace = []
i=0
for line in newdata:
    if (int(line[0]) == i and int(line[1])==1):
        freespace.append(finalVolume - int(line[3]))
        i = i+1   
```

```python
plt.plot(freespace)
plt.title("Free space vs time")
plt.xlabel("MCS")
plt.ylabel("Free volume")
plt.savefig("Free space Control")
```

```python
noofcells = []
count =0
for i in range(len(newdata)):
    if (int(newdata[i][1]) != 1):
        count = count+1
    else :
        noofcells.append(count+1)
        count=0
```

```python
plt.plot(noofcells)
plt.title("Number of cells vs time")
plt.xlabel("MCS")
plt.ylabel("Number of cells")
plt.savefig("Number of cells Control")
```

```python
volume = []
varvolume = []
for i in range(len(newdata)):
    if (int(newdata[i][1]) != 1 and i != 0):
        volume.append(int(newdata[i-1][2]))
    else :
        volume.append(int(newdata[i-1][2]))
        npvolume = np.array(volume)
        varvolume.append(npvolume.std())
        volume = []
npvolume = np.array(volume)
varvolume.append(npvolume.std())
del varvolume[0]
```

```python
plt.plot(varvolume, 'r')
plt.title("Volume heterogeneity vs time")
plt.xlabel("MCS")
plt.ylabel("Volume heterogeneity")
plt.savefig("Volume heterogeneity Control")
```

```python
area = []
vararea = []
for i in range(len(newdata)):
    if (int(newdata[i][1]) != 1 and i != 0):
        area.append(float(newdata[i-1][4]))
    else :
        area.append(float(newdata[i-1][4]))
        nparea = np.array(area)
        vararea.append(nparea.std())
        area = []
nparea = np.array(area)
vararea.append(nparea.std())
del vararea[0]
```

```python
plt.plot(vararea, 'r' )
plt.title("Surface heterogeneity vs time")
plt.xlabel("MCS")
plt.ylabel("Surface heterogeneity")
plt.savefig("Surface heterogeneity Control")
```
