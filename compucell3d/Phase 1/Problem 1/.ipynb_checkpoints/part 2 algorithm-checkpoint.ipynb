```python
import csv
f1 = open ("C:\Users\Anmol Mishra\CC3DWorkspace\Latest\Triangle\Demo.cc3d-2016-05-06-13-40-40-705000\data.csv" , "rb")
data = csv.reader(f1)
```

```python
newdata = []
for line in data:
        newdata.append(line)
```

```python
finalvolume = int(newdata[-1][3])
for i in newdata:
    if int(i[3])==finalvolume:
        tend = int(i[0])
        break
```

```python
tend
```

```python
datatend = []
datatend100 = []
datatend200 = []
for i in newdata:
    if int(i[0])==tend:
        datatend.append(int(i[2]))
    if int(i[0])==tend+100:
        datatend100.append(int(i[2]))
    if int(i[0])==tend+200:
        datatend200.append(int(i[2]))
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
import numpy as np
x = np.array(datatend)
unique, counts = np.unique(x, return_counts=True)
print unique
```

```python
plt.plot(unique, counts , 'r-')
plt.title('Plot for tend ')
plt.xlabel('Volume of Cell')
plt.ylabel('Number of Occurrences')
plt.savefig("tend triangle.png")
```

```python
x = np.array(datatend100)
unique100, counts100 = np.unique(x, return_counts=True)
print unique100
```

```python
plt.plot(unique100, counts100 , 'r-')
plt.title('Plot for tend + 100')
plt.xlabel('Volume of Cell')
plt.ylabel('Number of Occurrences')
plt.savefig("tend100 triangle.png")
```

```python
x = np.array(datatend200)
unique200, counts200 = np.unique(x, return_counts=True)
print unique200
```

```python
plt.plot(unique200, counts200 , 'r-')
plt.title('Plot for tend + 200')
plt.xlabel('Volume of Cell')
plt.ylabel('Number of Occurrences')
plt.savefig(".png")
```

```python
plt.plot(unique, counts , label = 'tend' , linewidth = 5)
plt.plot(unique100, counts100 , label = 'tend + 100' , linewidth = 5)
plt.plot(unique200, counts200 , label = 'tend + 200' , linewidth = 5)
plt.title('Number of cells vs volume, tend = point of saturation')
plt.xlabel('Volume of Cell')
plt.ylabel('Number of Occurrences')
plt.legend()
plt.savefig("plotall.png")
```
