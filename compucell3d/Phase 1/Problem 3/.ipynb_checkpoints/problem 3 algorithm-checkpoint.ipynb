```python
import csv
f1 = open ("C:\Users\Anmol Mishra\CC3DWorkspace\Problem 3\Circle 1\Demo.cc3d-2016-05-15-13-46-37-293000\data.csv" , "rb")
data1 = csv.reader(f1)
f2 = open ("C:\Users\Anmol Mishra\CC3DWorkspace\Problem 3\Circle 2\Demo.cc3d-2016-05-15-13-25-57-397000\data.csv" , "rb")
data2 = csv.reader(f2)
f3 = open ("C:\Users\Anmol Mishra\CC3DWorkspace\Problem 3\Circle 3\Demo.cc3d-2016-05-15-13-16-51-275000\data.csv" , "rb")
data3 = csv.reader(f3)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
newdata1 = []
for line in data1:
        newdata1.append(line)
newdata2 = []
for line in data2:
        newdata2.append(line)
newdata3 = []
for line in data3:
        newdata3.append(line)
```

```python
freespace1 = []
i=0
for line in newdata1:
    if (int(line[0]) == i and int(line[1])==1):
        freespace1.append(300000 - int(line[3]))
        i = i+1
freespace2 = []
i=0
for line in newdata2:
    if (int(line[0]) == i and int(line[1])==1):
        freespace2.append(300000 - int(line[3]))
        i = i+1
freespace3 = []
i=0
for line in newdata3:
    if (int(line[0]) == i and int(line[1])==1):
        freespace3.append(300000 - int(line[3]))
        i = i+1
```

```python
for i in range(1350):
    freespace1[i] = freespace1[i] - 130000 
    freespace2[i] = freespace2[i] - 130000 
    freespace3[i] = freespace3[i] - 130000 
```

```python
plt.plot(freespace1 , linewidth = 2 , label = "R")
plt.plot(freespace2 , linewidth = 2 , label = "2R")
plt.plot(freespace3 , linewidth = 2 , label = "R/2")
plt.title("Free space vs time")
plt.xlabel("MCS")
plt.ylabel("Free volume")
plt.legend()
plt.savefig("Free space Problem 3")
```

```python
noofcells1 = []
count =0
for i in range(len(newdata1)):
    if (int(newdata1[i][1]) != 1):
        count = count+1
    else :
        noofcells1.append(count+1)
        count=0
noofcells2 = []
count =0
for i in range(len(newdata2)):
    if (int(newdata2[i][1]) != 1):
        count = count+1
    else :
        noofcells2.append(count+1)
        count=0
noofcells3 = []
count =0
for i in range(len(newdata3)):
    if (int(newdata3[i][1]) != 1):
        count = count+1
    else :
        noofcells3.append(count+1)
        count=0
```

```python
plt.plot( noofcells1, linewidth = 2 , label = "Circle 1")
plt.plot( [931], noofcells1[931] , 'ro' , label = "T_end")
plt.plot( [931], noofcells1[931] , 'go', label = "Tau")
plt.title("Number of cells vs time")
plt.xlabel("MCS")
plt.ylabel("Number of cells")
plt.legend( loc = 2)
plt.savefig("Number of Cells Circle 1")
```

```python
plt.plot( noofcells2, linewidth = 2 , label = "Circle 2")
plt.plot( [931], noofcells2[931] , 'ro' , label = "T_end")
plt.plot( [1031], noofcells2[1031] , 'go', label = "Tau")
plt.title("Number of cells vs time")
plt.xlabel("MCS")
plt.ylabel("Number of cells")
plt.legend( loc = 2)
plt.savefig("Number of Cells Circle 2")
```

```python
plt.plot( noofcells3, linewidth = 2 , label = "Circle 3")
plt.plot ([1400], [9000])
plt.plot( [931], noofcells3[931] , 'ro' , label = "T_end")
plt.plot( [1131], noofcells3[1131] , 'go', label = "Tau")
plt.title("Number of cells vs time")
plt.xlabel("MCS")
plt.ylabel("Number of cells")
plt.legend(loc = 2)
plt.savefig("Number of Cells Circle 3")
```

```python
volume1 = []
varvolume1 = []
for i in range(len(newdata1)):
    if (int(newdata1[i][1]) != 1 and i != 0):
        volume1.append(int(newdata1[i-1][2]))
    else :
        volume1.append(int(newdata1[i-1][2]))
        npvolume1 = np.array(volume1)
        varvolume1.append(npvolume1.std())
        volume1 = []
npvolume1 = np.array(volume1)
varvolume1.append(npvolume1.std())
del varvolume1[0]
volume2 = []
varvolume2 = []
for i in range(len(newdata2)):
    if (int(newdata2[i][1]) != 1 and i != 0):
        volume2.append(int(newdata2[i-1][2]))
    else :
        volume2.append(int(newdata2[i-1][2]))
        npvolume2 = np.array(volume2)
        varvolume2.append(npvolume2.std())
        volume2 = []
npvolume2 = np.array(volume2)
varvolume2.append(npvolume2.std())
del varvolume2[0]
volume3 = []
varvolume3 = []
for i in range(len(newdata3)):
    if (int(newdata3[i][1]) != 1 and i != 0):
        volume3.append(int(newdata3[i-1][2]))
    else :
        volume3.append(int(newdata3[i-1][2]))
        npvolume3 = np.array(volume3)
        varvolume3.append(npvolume3.std())
        volume3 = []
npvolume3 = np.array(volume3)
varvolume3.append(npvolume3.std())
del varvolume3[0]
```
