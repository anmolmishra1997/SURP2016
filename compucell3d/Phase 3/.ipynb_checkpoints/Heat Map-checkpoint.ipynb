```python
import csv
f1 = open ("C:\Users\Anmol Mishra\CC3DWorkspace\Phase 3 New\Jcc =10\Circle R\Demo_cc3d_05_28_2016_18_59_55\data.csv" , "rb")
data = csv.reader(f1)
```

```python
newdata = []
for line in data:
        newdata.append(line)
```

```python
finalVolume = 0
for i in range(len(newdata)):
    if int(newdata[i][3])>finalVolume:
        finalVolume = int(newdata[i][3])
```

```python
tend = 0
for i in range(len(newdata)):
    if int(newdata[i][3])==finalVolume:
        tend = int(newdata[i][0])
        break
```

```python
tend
```

```python
x= []
y =[]
z = []
x100 = []
y100 = []
z100 = []
x200 = []
y200 = []
z200 = []
for i in newdata:
    if int(i[0])==tend:
        x.append(float(i[5]))
        y.append(float(i[6]))
        z.append(int(i[2]))
    if int(i[0])==tend+100:
        x100.append(float(i[5]))
        y100.append(float(i[6]))
        z100.append(int(i[2]))
    if int(i[0])==tend+200:
        x200.append(float(i[5]))
        y200.append(float(i[6]))
        z200.append(int(i[2]))
    if int(i[0])>tend +200:
        break
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate
%matplotlib inline
```

```python
# Generate data:
x = np.array(x)
y = np.array(y)
z = np.array(z)
# Set up a regular grid of interpolation points
xi, yi = np.linspace(x.min(), x.max(), 500), np.linspace(y.min(), y.max(), 500)
xi, yi = np.meshgrid(xi, yi)

# Interpolate
rbf = scipy.interpolate.Rbf(x, y, z, function='linear')
zi = rbf(xi, yi)

plt.imshow(zi, vmin=z.min(), vmax=z.max(), origin='lower',
           extent=[x.min(), x.max(), y.min(), y.max()])
plt.scatter(x, y, c=z)
plt.colorbar()
plt.show()
plt.savefig("Volume at tend")
```
